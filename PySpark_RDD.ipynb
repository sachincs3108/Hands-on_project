```python
# Loading and partitioning data
rdd_iris1 = sc.textFile("/FileStore/tables/Iris___Copy.csv")
```

```python
# print data 
rdd_iris1.collect()
```

```python
rdd_iris1.getNumPartitions()
```

```python
# show data in the 0th partition
rdd_iris1.glom().collect()[0]
```

```python
# No. of data in the 0th partition
len(rdd_iris1.glom().collect()[0])
```

```python
# No. of data in the 1st partion
len(rdd_iris1.glom().collect()[1])
```

```python
# Partitioning into 8 parts
rdd_iris2 = sc.parallelize(rdd_iris1.collect(),8)
```

```python
rdd_iris2.getNumPartitions()
```

```python
for i in rdd_iris1.collect():
    print(i)
```

```python
rdd_iris3 = rdd_iris2.map(lambda x: x*2)
```

```python
rdd_iris3.collect()
```

```python
rdd_iris4 = rdd_iris2.flatMap(lambda x: x.split(','))
```

```python
# view
rdd_iris4.collect()
```

```python
rdd_iris4.getNumPartitions()
```

```python
rdd_iris4.glom().collect()[0]
```

```python
rdd_iris4.glom().collect()[1]
```

```python
rdd_iris5 = rdd_iris4.map(lambda x: x*2)
```

```python
rdd_iris5.collect()
```

```python
# Filter the data accoring to required species
rdd_iris6 = rdd_iris1.flatMap(lambda x: x.split(",")).filter(lambda x: x=='Iris-setosa').map(lambda x:(x,1))
```

```python
rdd_iris6.glom().collect()[0]
```

```python
# narrow transformation
rdd_species = rdd_iris1.flatMap(lambda x: x.split(",")).filter(lambda x: (x=="Iris-virginica" or x=="Iris-setosa" or x=="Iris-versicolor"))
```

```python
rdd_species.collect()
```

```python
# group by
rdd_spec1 = rdd_species.groupBy(lambda x:x).mapValues(lambda x: (len(x)))
```

```python
rdd_spec1.collect()
```
